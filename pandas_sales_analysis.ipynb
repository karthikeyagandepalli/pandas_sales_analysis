<h4> EXPLORATORY DATA ANALYSIS

- <H4> what is the overall slaes trend

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
#getting month_year from the dataset
df['month_year']= df['order_date'].apply(lambda x: x.strftime('%y-%m'))
```

```python
# Groupinf month year
df_trend = df.groupby('month_year')['sales'].sum().reset_index()
```

```python
#setting the figure size 
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'],df_trend['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

- <h2> WHICH ARE THE TOP 10 PRODUCTS BY SALE?

```python
df.head()
```

```python
#Grouping product name cloumns
prod_sales = pd.DataFrame(df.groupby('product_name')['sales'].sum())
```

```python
#sorting product sales column
prod_sales = prod_sales.sort_values('sales',ascending=False)
```

```python
# Top 10 products by sales
prod_sales[:10]
```

- <H4> WHICH ARE THE MOST SELLING PRODUCTS?

```python
#Grouping product names
most_selling_prod = pd.DataFrame(df.groupby('product_name')['quantity'].sum())
```

```python
# sorting most_selling products
most_selling_prod = most_selling_prod.sort_values('quantity',ascending=False)
```

```python
most_selling_prod[:10]
```

- <H4> MOST PREFERRED SHIP MODE?

```python
#preferred ship mode
sns.countplot(df['ship_mode'])
```

- <h4> WHICH ARE THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY

```python
# Grouping category and subcategory
cat_subcat_profit = pd.DataFrame(df.groupby(['category','sub_category'])['profit'].sum())
```

```python
#sorting the results
cat_subcat_profit.sort_values(['category','profit'], ascending = False)
```
